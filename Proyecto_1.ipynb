**Ejercicio N°1:**


Escribir un programa que pregunte al usuario por las ventas de los últimos 5 días, y muestre por pantalla una serie con los datos de las ventas por producto: Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.


```python
print("Ingrese la venta por producto del dia t-1  ")
dia1 = int(input())
print("Ingrese la venta por producto del dia t-2  ")
dia2 = int(input())
print("Ingrese la venta por producto del dia t-3  ")
dia3 = int(input())
print("Ingrese la venta por producto del dia t-4  ")
dia4 = int(input())
print("Ingrese la venta por producto del dia t-5  ")
dia5 = int(input())
print("Ingrese el precio unitario  ")
precio = float(input())
```

Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.


```python
dic = {} #Creamos nuestro diccionario
```

```python
suma = sum(dic.values()) # calculamos la suma de los valores del diccionario
tamano = len (dic.keys()) # obtenemos el numero de keys
avg = round(suma/tamano,2) #Calculamos el promedio de los valores del diccionario
print("El salario minino es ", min(dic.values()))
print("El salario maximo es ", max(dic.values()))
print("El promedio de los salarios es", avg)
```

```python
En = ["Enero",30500,22000]
Fe = ["Febrero",35600,23450]
Ma = ["Marzo",28300,18100]
Ab = ["Abril",33900,35700]
May = ["Mayo",42500,32450]

Lst = [En,Fe,Ma,Ab,May] #Ingresamos los valores de la tabla a una lista

df = pd.DataFrame(Lst, columns = ["Mes", "Gastos", "Ventas"]) #Creamos el Data Drame
df
```

```python
df["Columna1"]= df['Gastos']/df['Ventas']*100 #Calculamos los valores de la nueva columna
df
```

```python
df["Columna2"] = np.where(df["Ventas"]>=30000,"Meta superada","Meta no superada") #Calculamos los valores de la nueva columna
                                                                                               
df
```

```python
df["Columna3"] = np.where(df["Gastos"]>df["Ventas"], df["Gastos"] - df["Ventas"],"-") #Calculamos los valores de la nueva columna
                                                                                               
df
```

```python
Ventas_um = df['Ventas'][2]+df['Ventas'][3]+df['Ventas'][4] #Obtenemos el total de las ventas de los últimos 3 meses
Ventas_um
```

```python
Gastos_tm = df["Gastos"][1]+df["Gastos"][2]+df["Gastos"][3] #Obtenemos el total de gastos de febrero a abril
Gastos_tm
```
